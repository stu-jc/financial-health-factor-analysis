### Load Data
**Load the data from the provided CSV file and display the first few rows to inspect the data structure.**

```python
# Load the data from the provided file
INTCOV = # Load dataset
pd.read_csv("INTCOV.zip",parse_dates=["date"]).set_index(["id","date"]).dropna()

OCF_LCT = # Load dataset
pd.read_csv("OCF_LCT.zip",parse_dates=["date"]).set_index(["id","date"]).dropna()

Cash_Debt = # Load dataset
pd.read_csv("CASH_DEBT.zip",parse_dates=["date"]).set_index(["id","date"]).dropna()

gp = # Load dataset
pd.read_csv("GP.zip",parse_dates=["date"]).set_index(["id","date"]).dropna()

mom12 = # Load dataset
pd.read_csv("MOM12.zip",parse_dates=["date"]).set_index(["id","date"]).dropna()

TVOL = # Load dataset
pd.read_csv("TVOL.zip",parse_dates=["date"]).set_index(["id","date"]).dropna()

EPQ = # Load dataset
pd.read_csv("EPQ.zip",parse_dates=["date"]).set_index(["id","date"]).dropna()

prices = # Load dataset
pd.read_csv("Prices.zip",parse_dates=["date"]).set_index(["id","date"]).dropna()

market_cap = # Load dataset
pd.read_csv("Market_cap.zip",parse_dates=["date"]).set_index(["id","date"]).dropna()

Names = # Load dataset
pd.read_csv("names.zip").set_index("id")

Macro = # Load dataset
pd.read_csv("INT_RATES.csv").set_index("date")
```

**Signals**

```python
# Execute the following code
# Remove rows with zero or negative prices
prices 

# Setting our signal to the Earnings to Price Ratio ("EPQ") to derive the information coefficient
signal = EPQ.join([INTCOV, OCF_LCT, Cash_Debt], how='inner')

# Handle missing or infinite values in the signal data
signal = signal.replace([np.inf, -np.inf], np.nan)
signal = signal.dropna()

signal.head()
```

**Future Returns**

```python
# Execute the following code
future_returns = np.log(prices['price'].groupby('id').shift(-1) / prices['price']).rename('fut_ret')
future_returns.head()
```

**Calculating IC Factors**

```python
# Execute the following code
# Calculate IC for each factor (monthly):
print("IC for each factor (monthly):")
ic_inctcov_monthly = ap.ic_analysis(signal['intcov'], prices['price'], frequency='monthly')
ic_ocf_lct_monthly = ap.ic_analysis(signal['ocf_lct'], prices['price'], frequency='monthly')
ic_cash_debt_monthly = ap.ic_analysis(signal['cash_debt'], prices['price'], frequency='monthly')

# Calculate IC for each factor (quarterly):
print("IC for each factor (quarterly):")
ic_inctcov_quarterly = ap.ic_analysis(signal['intcov'], prices['price'], frequency='quarterly')
ic_ocf_lct_quarterly = ap.ic_analysis(signal['ocf_lct'], prices['price'], frequency='quarterly')
ic_cash_debt_quarterly = ap.ic_analysis(signal['cash_debt'], prices['price'], frequency='quarterly')

# Calculate IC for each factor (annual):
print("IC for each factor (annual):")
ic_inctcov_annual = ap.ic_analysis(signal['intcov'], prices['price'], frequency='annual')
ic_ocf_lct_annual = ap.ic_analysis(signal['ocf_lct'], prices['price'], frequency='annual')
ic_cash_debt_annual = ap.ic_analysis(signal['cash_debt'], prices['price'], frequency='annual')
```

**Average Return Calculations**

```python
# Execute the following code
# Create a synthetic "Average" factor (monthly)
avg_factor_monthly = (signal['intcov'] + signal['ocf_lct'] + signal['cash_debt']) / 3
avg_factor_monthly.name = 'AVG_FACTOR'
print("Synthetic IC (monthly):")
ic_avg_monthly = ap.ic_analysis(avg_factor_monthly, prices['price'], frequency='monthly')

# Create a synthetic "Average" factor (quarterly)
avg_factor_quarterly = (signal['intcov'] + signal['ocf_lct'] + signal['cash_debt']) / 3
avg_factor_quarterly.name = 'AVG_FACTOR'
print("Synthetic IC (quarterly):")
ic_avg_quarterly = ap.ic_analysis(avg_factor_quarterly, prices['price'], frequency='quarterly')

# Create a synthetic "Average" factor (annual)
avg_factor_annual = (signal['intcov'] + signal['ocf_lct'] + signal['cash_debt']) / 3
avg_factor_annual.name = 'AVG_FACTOR'
print("Synthetic IC (annual):")
ic_avg_annual = ap.ic_analysis(avg_factor_annual, prices['price'], frequency='annual')
```

**DataFrame Creation relevant to IC series**

```python
# Execute the following code
# Create DataFrames with all IC series
ic_df_monthly = pd.DataFrame({'IC_INCTCOV': ic_inctcov_monthly, 
                              'IC_OCF_LCT': ic_ocf_lct_monthly, 
                              'IC_CASH_DEBT': ic_cash_debt_monthly, 
                              'IC_AVG': ic_avg_monthly})

ic_df_quarterly = pd.DataFrame({'IC_INCTCOV': ic_inctcov_quarterly, 
                                'IC_OCF_LCT': ic_ocf_lct_quarterly, 
                                'IC_CASH_DEBT': ic_cash_debt_quarterly, 
                                'IC_AVG': ic_avg_quarterly})

ic_df_annual = pd.DataFrame({'IC_INCTCOV': ic_inctcov_annual, 
                             'IC_OCF_LCT': ic_ocf_lct_annual, 
                             'IC_CASH_DEBT': ic_cash_debt_annual, 
                             'IC_AVG': ic_avg_annual})
```

```python
# Execute the following code
# I have split the data so that the sample (where we will test our data) is in the required range 
# and the training data is the earliest date available in db2 after we have cleared it of not available (NA)
# data to 2006

X_train = X.loc[pd.IndexSlice[:, '1996-07-31':'2006-12-31'],:]
X_test = X.loc[pd.IndexSlice[:, '2007-01-01':'2023-12-31'],:]

y_train = y.loc[pd.IndexSlice[:, '1996-07-31':'2006-12-31']]
y_test = y.loc[pd.IndexSlice[:, '2007-01-01':'2023-12-31']]
```

**1.** We’ve confirmed that our model is compatible with our data signals. The next step is to evaluate its performance.

    Accuracy Rate

**3.** This can be achieved by evaluating the model on both in-sample data (training set) and out-of-sample data (test set).

We will first generate predictions and then measure accuracy using the metrics module from sklearn, which provides score functions, performance metrics, and other useful evaluation tools.

### Import Libraries
**Import necessary libraries and modules for the analysis.**

```python
# Import necessary libraries and modules
# In-sample predictions
y_train_pred = clf.predict(X_train)

# Out-of-sample predictions
y_test_pred = clf.predict(X_test)

# We load the metrics module
from sklearn import metrics

# We measure the accuracy
is_accuracy = metrics.accuracy_score(y_train, y_train_pred)
os_accuracy = metrics.accuracy_score(y_test, y_test_pred)

# We format and print the results
print(f'In-Sample Accuracy: {round(is_accuracy, 3)}') # The round function is to format the number
print(f'Out-of-Sample Accuracy:, {round(os_accuracy, 3)}')
```

Our In-Sample and Out-of-Sample Accuracy is similar, so we can be relatively certain that our model performs equally well in either samples. 

To further validate our accuracy, we can perform a cross-validation model where we can repeat the train-test procedure with different splits of our sample. 

This essentially means that the training data will be split in 10 different ways to cross validate the training to give it more variaty in the same set of training data. 

### Import Libraries
**Import necessary libraries and modules for the analysis.**

```python
# Import necessary libraries and modules
from sklearn.model_selection import cross_val_score

scores = cross_val_score(clf, X, y, cv=10)
print(f'10-Folds Cross Validation Average Accuracy:, {round(scores.mean(), 3)}')
```

### Cross-Validation
**Perform cross-validation to evaluate the model's performance.**

```python
# Perform cross-validation to evaluate model performance
# Assuming X_train and y_train are your training data

# Create an empty list to store the results
optimal = []

# Start the loop
for depth in range(1, 8):
    # Instantiate the classifier
    clf = tree.DecisionTreeClassifier(max_depth=depth)
    
    # Run the 10-fold cross-validation on the train set only
    kscores = cross_val_score(clf, X_train, y_train, cv=10)       
    
    # Save the average accuracy (and the number of neighbors)
    optimal.append([depth, kscores.mean()])

# Transform the results into a DataFrame
optimal = pd.DataFrame(optimal, columns=['Tree Depth', 'Accuracy']).set_index('Tree Depth')

# Plot
optimal.plot(figsize=(8, 5), 
             style='o-',  # Change to circle markers with solid lines
             color='orange',  # Change the color to orange
             xlabel='Tree Depth',
             ylabel='10-Folds Average Out-of-Sample Accuracy',
             title='Hyperparameter Tuning')

plt.show()
```

```python
# Execute the following code
opt_depth = optimal['Accuracy'].idxmax()
opt_depth
```
